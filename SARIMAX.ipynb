```python
num_of_trials = [0, 3, 10, 20, 50,100]
data = stats.bernoulli.rvs(0.5, size = 100)
x = np.linspace(0,1, 100)

for i,N in enumerate(num_of_trials):
    heads = data[:N].sum()
    
    ax = plt.subplot(len(num_of_trials)/2, 2, i+1)
    ax.set_title('{} trials, {} heads'.format(N, heads))
    if i == 0:
        plt.ylim([0.0, 2.0])
    y = stats.beta.pdf(x, heads+1, 1+N - heads)
    plt.plot(x, y, label='{} tosses, {} heads'.format(N, heads))
    plt.fill_between(x, 0, y, alpha=0.5)
    
plt.tight_layout()
```

```python
air_pass = pd.read_csv('AirPassengers.csv', index_col='Month')
air_pass.index = pd.to_datetime(air_pass.index)
air_pass.head()
```

```python
air_pass['first_diff'] = air_pass['season_diff'] - air_pass['season_diff'].shift(1)
air_pass.dropna(inplace=True)
with plt.style.context('ggplot'):
    plt.figure(figsize=(10,5))
    plt.plot(air_pass.first_diff)
print("Критерий Дики-Фуллера (стационарность ряда): {}".format(smt.stattools.adfuller(air_pass['first_diff'].values)[1]))
```

```python
smt.graphics.plot_acf(air_pass['first_diff'])
smt.graphics.plot_pacf(air_pass['first_diff'])
plt.show()
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')

results_arima = []
best_model = None
best_params = (0,0,0)
best_aic = float('inf')

for P in range(1,5):
    for Q in range(1,5):
        for p in range(1,5):
            for q in range(1,5):

                try:
                    arima_result = smt.statespace.SARIMAX(air_pass['#Passengers'].dropna(), order=(p,1,q), seasonal_order=(P,1,Q,12)).fit(disp=-1)
                    results_arima.append([arima_result.aic, (p,q,P,Q)])
                    print("Оценена модель: SARIMA{}".format((p,q,P,Q)))
                    if arima_result.aic < best_aic:
                        best_aic = arima_result.aic
                        best_params = (p,q,P,Q)
                        best_model = arima_result
                    
                except:
                    print('Эти параметры выдают ошибку: {}'.format((p,q,P,Q)))
```

```python
results_arima_df = pd.DataFrame(results_arima, columns=['AIC', 'Параметры модели'])
results_arima_df.sort_values(by='AIC').head()
```
